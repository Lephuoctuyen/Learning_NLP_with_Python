## Combining similar words – lemmatization

A similar technique to stemming is lemmatization. The difference is that lemmatization provides us with a real word, that is, its canonical form. For example, the lemma of the word cats is cat, and the lemma for the word ran is run

```python
from nltk.probability import FreqDist
def compile_stopwords_list_frequency(text, cut_off=0.02):
    words = tokenize_nltk(text)
    freq_dist = FreqDist(word.lower() for word in words)
    words_with_frequencies = [(word, freq_dist[word]) for word in freq_dist.keys()]
    sorted_words = sorted(words_with_frequencies, key=lambda tup: tup[1])
    stopwords = []
    if (type(cut_off) is int):
        stopwords = [tuple[0] for tuple in sorted_words if tuple[1] > cut_off] # First option: use a frequency cutoff
    elif (type(cut_off) is float):
        length_cutoff = int(cut_off*len(sorted_words)) # Second option: use a percentage of the words
        stopwords = [tuple[0] for tuple in sorted_words[-length_cutoff:]]
    else:
        raise TypeError("The cut off needs to be either a float (percentage) or an int (frequency cut off)")
    return stopwords
text = read_text_file("/content/sherlock_holmes_1.txt")
stopwords = compile_stopwords_list_frequency(text)
print(stopwords)
print(len(stopwords))

text = read_text_file("/content/sherlock_holmes_1.txt")
stopwords = compile_stopwords_list_frequency(text, cut_off=0.05)
print(stopwords)
print(len(stopwords))

stopwords = compile_stopwords_list_frequency(text, cut_off=100)
print(stopwords)
print(len(stopwords))
```

```python
from enum import Enum

class Noun_number(Enum):
  SINGULAR = 1
  PLURAL = 2

def get_nouns_number(text, model, method='lemma'):
  nouns = []
  doc = model(text)
  for token in doc:
    if (token.pos_ == 'NOUN'):
      if method == 'lemma':
        if token.lemma_ != token.text:
            nouns.append((token.text, Noun_number.PLURAL))
        else:
            nouns.append((token.text, Noun_number.SINGULAR))
      elif method == 'morph':
        if token.morph.get("Number") == "sing":
            nouns.append((token.text, Noun_number.PLURAL))
        else:
            nouns.append((token.text, Noun_number.SINGULAR))
  return nouns
```

```python
from anytree import Node, RenderTree
import spacy

# Tải mô hình spaCy
nlp = spacy.load("en_core_web_sm")  # Hoặc thay bằng mô hình tương ứng cho ngôn ngữ bạn dùng

def build_tree(token, parent_node=None):
    """Xây dựng cây từ token của spaCy."""
    current_node = Node(token.text, parent=parent_node)
    for child in token.children:  # Duyệt qua các từ con (children)
        build_tree(child, current_node)
    return current_node

def visualize_tree(sentence, model):
    doc = model(sentence)
    root = None
    for token in doc:
        if token.head == token:  # Tìm gốc (root) của cây cú pháp
            root = build_tree(token)
            break

    # Hiển thị cây
    if root:
        print(f"Cây cú pháp cho câu: {sentence}")
        for pre, _, node in RenderTree(root):
            print(f"{pre}{node.name}")
    else:
        print("Không thể tạo cây cú pháp.")

# Ví dụ sử dụng
sentences = ["I like programming in Python.", "This is an example of syntax parsing."]
for sentence in sentences:
    visualize_tree(sentence, nlp)
```

## Splitting sentences into clauses

### **Tách câu thành các mệnh đề**

Khi làm việc với văn bản, chúng ta thường gặp các câu ghép (compound sentences - các câu có hai phần quan trọng như nhau) và câu phức (complex sentences - câu có một phần phụ thuộc vào phần khác). Việc tách các câu ghép/phức này thành các mệnh đề thành phần có thể giúp xử lý dễ dàng hơn ở các bước tiếp theo. Công thức này sử dụng dependency parse để thực hiện điều đó.

---

### **Cách hoạt động**
- Cách mã hoạt động dựa trên cấu trúc của câu phức và câu ghép:
  - Mỗi mệnh đề chứa một động từ, và một trong số đó là động từ chính của câu (*root*).
  - Mã tìm động từ *root* (được đánh dấu bằng thẻ *ROOT* trong xử lý bằng spaCy), sau đó tìm các động từ khác trong câu.
  - Sử dụng thông tin về các từ con (*children*) của từng động từ để xác định ranh giới bên trái và phải của mệnh đề. Từ đó, mã dựng lại nội dung các mệnh đề.

---

### **Các bước thực hiện**

1. **Nhập thư viện và khởi tạo spaCy**:
   - Nhập gói spaCy.
   - Tải *spacy engine*.

2. **Xử lý câu ví dụ**:
   - Gán câu ví dụ vào biến `sentence`.
   - Sử dụng *spacy engine* để xử lý câu này.

3. **In thông tin dependency parse**:
   - Kết quả này sẽ giúp xác định cách tách câu thành các mệnh đề.

4. **Tìm động từ chính của câu**:
   - Định nghĩa hàm `find_root_of_sentence` để tìm token có thẻ *ROOT*.
   - Áp dụng hàm này để lấy động từ chính của câu.

5. **Tìm các động từ khác**:
   - Định nghĩa hàm `find_other_verbs` để tìm các động từ trong câu:
     - Các động từ này có thẻ *VERB* và chỉ có động từ *root* là tổ tiên của nó.
   - Áp dụng hàm này để xác định danh sách các động từ phụ.

6. **Xác định ranh giới mệnh đề cho mỗi động từ**:
   - Định nghĩa hàm `get_clause_token_span_for_verb` để tìm chỉ số bắt đầu và kết thúc cho mệnh đề của từng động từ.
   - Dựa trên chỉ số từ con trái nhất (bắt đầu) và phải nhất (kết thúc) của động từ đó.

7. **Tạo danh sách các mệnh đề**:
   - Sử dụng danh sách chỉ số từ ở bước trên để tạo các đối tượng `Span` (đại diện cho mệnh đề) từ câu gốc.
   - Đảm bảo các mệnh đề trong danh sách được sắp xếp theo thứ tự xuất hiện trong câu.

8. **In các mệnh đề**:
   - Kết quả cuối cùng là các mệnh đề của câu được in ra.
   - Lưu ý: Một số từ, như *but*, có thể bị thiếu nếu nó phụ thuộc vào động từ chính (*eats*). Việc xử lý bổ sung để bao gồm từ như *but* được để lại như một bài tập mở rộng.

```python
import spacy

compound_sentence = "He eats cheese, but he won’t eat ice cream."
complex_sentence = "If it rains later, we won't be able to go to the park."
nlp = spacy.load('en_core_web_sm')

verb_pos_set = ["VERB"]

# để tìm token có thẻ ROOT.
def find_root_of_sentence(doc):
    root_token = None
    for token in doc:
        if (token.dep_ == "ROOT"):
            root_token = token
    return root_token

def find_other_verbs(doc, root_token):
    other_verbs = []
    for token in doc:
        ancestors = list(token.ancestors)
        if (token.pos_ == "VERB" and len(ancestors) == 1 and ancestors[0] == root_token):
            other_verbs.append(token)
    return other_verbs

def get_clause_token_span_for_verb(verb, doc, all_verbs):
    first_token_index = len(doc)
    last_token_index = 0
    this_verb_children = list(verb.children)
    for child in this_verb_children:
        if (child not in all_verbs):
            if (child.i < first_token_index):
                first_token_index = child.i
            if (child.i > last_token_index):
                last_token_index = child.i

    return(first_token_index, last_token_index)

def potential_clause_contains_subj(clause):
    contains_subj = False
    for token in clause:
        if (token.dep_ == "nsubj"):
            contains_subj = True
    return contains_subj

def print_doc_info(doc):
    for token in doc:
        ancestors = [t.text for t in token.ancestors]
        children = [t.text for t in token.children]
        print(token.text, "\t", token.i, "\t", token.pos_, "\t", token.dep_, "\t", ancestors, "\t", children)

def get_clauses(sentence):
    doc = nlp(sentence)
    print_doc_info(doc)
    # Find the root token
    root_token = find_root_of_sentence(doc)
    print(root_token)
    # Find the other verbs
    other_verbs = find_other_verbs(doc, root_token)
    print(other_verbs)
    token_spans = []
    # Find the token span for each of the other verbs
    all_verbs = [root_token] + other_verbs
    for other_verb in all_verbs:
        (first_token_index, last_token_index) = get_clause_token_span_for_verb(other_verb, doc, all_verbs)
        token_spans.append((first_token_index, last_token_index))
    sentence_clauses = []
    print(token_spans)
    is_CCONJ = False
    for token in doc:
        if (token.pos_ == "CCONJ"):
            is_CCONJ = True

    for i, token_span in enumerate(token_spans):
        start, end = token_span
        if i > 0 and is_CCONJ:
            start -= 1
        if start < end:
            clause = doc[start:end]
            if potential_clause_contains_subj(clause):
                sentence_clauses.append(clause)

    sentence_clauses = sorted(sentence_clauses, key=lambda tup: tup[0])
    return sentence_clauses

def main():
    # clauses = get_clauses(complex_sentence)
    # clauses_text = [clause.text for clause in clauses]
    # print(clauses_text)
    clauses = get_clauses(compound_sentence)
    clauses_text = [clause.text for clause in clauses]
    print(clauses_text)


if (__name__ == "__main__"):
    main()
```

```python
def get_subject_phrase(doc):
    for token in doc:
        if ("subj" in token.dep_):
            subtree = list(token.subtree)
            start = subtree[0].i
            end = subtree[-1].i + 1
            return doc[start:end]
```

```python
def get_object_phrase(doc):
    for token in doc:
        if ("dobj" in token.dep_):
            subtree = list(token.subtree)
            start = subtree[0].i
            end = subtree[-1].i + 1
            return doc[start:end]
sentences = [
    "The big black cat stared at the small dog.",
    "Jane watched her brother in the evenings.",
    "Laura gave Sam a very interesting book."
]
for sentence in sentences:
    doc = nlp(sentence)
    subject_phrase = get_subject_phrase(doc)
    object_phrase = get_object_phrase(doc)
    print(sentence)
    print("\tSubject:", subject_phrase)
    print("\tDirect object:", object_phrase)
```

### Get dative phrase


```python
def get_dative_phrase(doc):
    for token in doc:
        if ("dative" in token.dep_):
            subtree = list(token.subtree)
            start = subtree[0].i
            end = subtree[-1].i + 1
            return doc[start:end]
```

### Combined function


```python
def get_phrase(doc, phrase):
    # phrase is one of "subj", "obj", "dative"
    for token in doc:
        if (phrase in token.dep_):
            subtree = list(token.subtree)
            start = subtree[0].i
            end = subtree[-1].i + 1
            return doc[start:end]
```

### Get prepositional phrases


```python
def get_prepositional_phrase_objs(doc):
    prep_spans = []
    for token in doc:
        if ("pobj" in token.dep_):
            subtree = list(token.subtree)
            start = subtree[0].i
            end = subtree[-1].i + 1
            prep_spans.append(doc[start:end])
    return prep_spans
sentence = "Laura gave Sam a very interesting book."
doc = nlp(sentence)
subject_phrase = get_phrase(doc, "subj")
object_phrase = get_phrase(doc, "obj")
dative_phrase = get_phrase(doc, "dative")
print(sentence)
print("\tSubject:", subject_phrase)
print("\tDirect object:", object_phrase)
print("\tDative object:", dative_phrase)
```

Tf-idf (Term Frequency-Inverse Document Frequency) là một phương pháp dùng để cân nhắc tầm quan trọng của các từ trong tài liệu hoặc từ vựng trong tập dữ liệu lớn. Nội dung chính của tf-idf bao gồm:

1. **Term Frequency (TF)**: Đo lường tần suất xuất hiện của từ trong tài liệu. Thường được điều chỉnh bằng logarit để giảm sự ảnh hưởng của những từ xuất hiện quá nhiều lần.
   
2. **Inverse Document Frequency (IDF)**: Đo lường mức độ phổ biến của từ trong toàn bộ tập dữ liệu. Các từ xuất hiện ở nhiều tài liệu sẽ có trọng số thấp hơn, trong khi các từ xuất hiện ít sẽ có trọng số cao hơn.

Tf-idf là tích của TF và IDF, giúp loại bỏ ảnh hưởng của các từ phổ biến nhưng không phân biệt được giữa các tài liệu. Ví dụ, từ "good" xuất hiện ở tất cả các vở kịch của Shakespeare nên bị loại bỏ, trong khi từ "Romeo" có trọng số cao hơn vì nó chỉ xuất hiện trong một vở.

Tf-idf được sử dụng rộng rãi trong các nhiệm vụ xử lý ngôn ngữ tự nhiên và truy xuất thông tin để làm giảm ảnh hưởng của các từ không quan trọng và làm nổi bật các từ mang tính phân biệt.

## Overview Levenshtein distance

**Khoảng cách chỉnh sửa tối thiểu** (minimum edit distance) và cách nó được tính toán bằng thuật toán lập trình động. Đây là một khái niệm quan trọng trong **xử lý ngôn ngữ tự nhiên** (NLP) và được áp dụng trong nhiều bài toán như:

1. **Chỉnh sửa lỗi chính tả**: Xác định từ nào gần giống nhất với từ bị sai (ví dụ: *graffe* gần với *giraffe*).
2. **Coreference Resolution**: Đánh giá xem hai chuỗi văn bản có tham chiếu cùng một thực thể hay không.
3. **Tính toán khoảng cách chuỗi**: Xác định số lượng các thao tác nhỏ nhất (thêm, xóa, thay thế ký tự) để biến một chuỗi thành chuỗi khác.

- **Khoảng cách chỉnh sửa** được tính bằng cách xác định số thao tác:
   - **Insertion** (thêm ký tự)
   - **Deletion** (xóa ký tự)
   - **Substitution** (thay thế ký tự)
- **Levenshtein Distance** là một phiên bản cụ thể trong đó:
   - Chi phí cho **insertion** và **deletion** là 1.
   - Chi phí cho **substitution** là 2 (trừ khi hai ký tự giống nhau thì chi phí là 0).

- Thuật toán tính **khoảng cách chỉnh sửa tối thiểu** sử dụng **lập trình động** (dynamic programming) để tối ưu hóa việc tính toán.
   - Bước cơ sở: Chi phí cho các chuỗi rỗng.
   - Quan hệ đệ quy: Sử dụng giá trị tối thiểu từ các trạng thái trước đó.

Công thức tính:
$$
D[i, j] = \min \begin{cases}
D[i-1, j] + 1 & \text{(xóa ký tự)} \\
D[i, j-1] + 1 & \text{(thêm ký tự)} \\
D[i-1, j-1] + 2 & \text{(thay thế nếu ký tự khác nhau)} \\
D[i-1, j-1] + 0 & \text{(nếu ký tự giống nhau)}
\end{cases}
$$

 Ứng dụng:
- **Speech Recognition**: Căn chỉnh các chuỗi văn bản (alignment).
- **Machine Translation**: So khớp các chuỗi giữa ngôn ngữ nguồn và đích.
- **DNA Sequence Alignment**: So sánh các chuỗi DNA để tìm đột biến.

 Ví dụ minh họa:
- Biến chuỗi `intention` thành `execution` với chi phí tối thiểu:
   - Các thao tác bao gồm: xóa, thêm và thay thế ký tự.
   - Sử dụng lập trình động để tính toán khoảng cách một cách hiệu quả.

In this recipe, we will use two simple tools for labeling a sentence as having positive or negative sentiment. The first tool is the NLTK Vader sentiment analyzer, and the second one uses the textblob package

**VADER (Valence Aware Dictionary and sEntiment Reasoner)** là công cụ phân tích cảm xúc dựa trên từ điển và quy tắc, đặc biệt nhạy cảm với cảm xúc trên các nền tảng truyền thông xã hội. VADER sử dụng từ vựng đánh giá cảm xúc tích cực hay tiêu cực và mức độ mạnh mẽ của cảm xúc. Công cụ trả về bốn giá trị chính: **neg** (tiêu cực), **neu** (trung lập), **pos** (tích cực), và **compound** (tổng hợp, chuẩn hóa từ -1 đến 1).

VADER đặc biệt hiệu quả trong việc xử lý ngôn ngữ không chính thống trên mạng xã hội, bao gồm:
1. **Biến thể ngữ pháp** (viết tắt, ký hiệu, từ lóng).
2. **Biểu cảm cảm xúc** (biểu tượng cảm xúc).
3. **Từ ngữ mơ hồ** (mỉa mai, trào phúng).
4. **Cảm xúc mạnh mẽ/phóng đại** (như "hated", "love").
5. **Hashtags và mentions** (ví dụ: #love, @username).
6. **Từ ngữ liên quan đến sự kiện thời sự** (như "climate change", "election").

Với khả năng phân tích cảm xúc trong các yếu tố phức tạp và ngôn ngữ không chính thức, VADER là công cụ hữu ích trong việc phân tích cảm xúc trên mạng xã hội.

**Văn bản không chính thức** (informal text) là những văn bản mang tính thân mật, suồng sã, hoặc không tuân theo các quy tắc ngữ pháp nghiêm ngặt. Nó thường xuất hiện trong giao tiếp hàng ngày, tin nhắn, mạng xã hội, hoặc hội thoại giữa bạn bè và người thân. Ví dụ, các câu trong văn bản không chính thức có thể sử dụng từ ngữ lóng, viết tắt, hoặc cú pháp đơn giản.

Ví dụ:
- "What's up?" (Thân mật, viết tắt)
- "I luv this!" (Viết tắt, dùng từ lóng)

**Positive sentiment**: Thể hiện sự tích cực (e.g., "Great job!").  
**Negative sentiment**: Thể hiện sự tiêu cực (e.g., "Terrible work.").
